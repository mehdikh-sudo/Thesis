## Feature Diagnostic Script

```python
"""
Feature Diagnostic Script
Run this BEFORE training to catch silent bugs in your extracted features.
Checks: I3D RGB, I3D Flow, ResNet features for SumMe/TVSum
"""

import numpy as np
import h5py
import os
from collections import defaultdict

# ── CONFIG ──────────────────────────────────────────────────────────────────
# Change these paths to match your setup
FEATURE_FILE = out_h5_path  # or .npy / .npz
DATASET_NAME = "SumMe"              # "SumMe" or "TVSum"

# ── 1. BASIC SHAPE & DTYPE ──────────────────────────────────────────────────
def check_shapes(features_dict):
    """
    features_dict = {
        'video_name': {
            'resnet': np.array [N, 512],
            'i3d_rgb': np.array [N, 1024],
            'i3d_flow': np.array [N, 1024],
            'concat_pca': np.array [N, 512],  # after PCA
            'gtscore': np.array [N],           # ground truth importance
        }
    }
    """
    print("=" * 60)
    print("CHECK 1: SHAPES & DTYPES")
    print("=" * 60)

    for vid, feats in features_dict.items():
        resnet = feats['resnet']
        rgb    = feats['i3d_rgb']
        flow   = feats['i3d_flow']
        pca    = feats['concat_pca']
        gt     = feats['gtscore']

        issues = []

        # Shape checks
        N = resnet.shape[0]
        if rgb.shape[0] != N:
            issues.append(f"I3D RGB has {rgb.shape[0]} frames, ResNet has {N}")
        if flow.shape[0] != N:
            issues.append(f"I3D Flow has {flow.shape[0]} frames, ResNet has {N}")
        if pca.shape[0] != N:
            issues.append(f"PCA has {pca.shape[0]} frames, ResNet has {N}")
        if gt.shape[0] != N:
            issues.append(f"GT score has {gt.shape[0]} values, ResNet has {N}")

        # Dim checks
        if resnet.shape[1] != 512:
            issues.append(f"ResNet dim={resnet.shape[1]}, expected 512")
        if rgb.shape[1] != 1024:
            issues.append(f"I3D RGB dim={rgb.shape[1]}, expected 1024")
        if flow.shape[1] != 1024:
            issues.append(f"I3D Flow dim={flow.shape[1]}, expected 1024")
        if pca.shape[1] != 512:
            issues.append(f"PCA dim={pca.shape[1]}, expected 512")

        # Dtype
        for name, arr in [('resnet', resnet), ('i3d_rgb', rgb), ('i3d_flow', flow), ('pca', pca)]:
            if arr.dtype not in [np.float32, np.float64]:
                issues.append(f"{name} dtype={arr.dtype}, should be float32/64")

        status = "❌ ISSUES" if issues else "✓ OK"
        print(f"\n{vid}: {status}")
        print(f"  Frames: {N} | ResNet:{resnet.shape} RGB:{rgb.shape} Flow:{flow.shape} PCA:{pca.shape}")
        for iss in issues:
            print(f"  ⚠️  {iss}")


# ── 2. NaN / INF CHECK ──────────────────────────────────────────────────────
def check_nan_inf(features_dict):
    print("\n" + "=" * 60)
    print("CHECK 2: NaN / Inf VALUES")
    print("=" * 60)

    any_issue = False
    for vid, feats in features_dict.items():
        for name, arr in feats.items():
            if isinstance(arr, np.ndarray) and np.issubdtype(arr.dtype, np.floating):
                n_nan = np.isnan(arr).sum()
                n_inf = np.isinf(arr).sum()
                if n_nan > 0 or n_inf > 0:
                    print(f"❌ {vid} / {name}: {n_nan} NaN, {n_inf} Inf")
                    any_issue = True

    if not any_issue:
        print("✓ No NaN or Inf values found")


# ── 3. VALUE RANGE & SCALE ──────────────────────────────────────────────────
def check_scales(features_dict):
    """
    CRITICAL: ResNet (ImageNet) and I3D (Kinetics) features
    live in very different numerical ranges.
    PCA will be dominated by whichever has larger variance.
    """
    print("\n" + "=" * 60)
    print("CHECK 3: VALUE RANGES & SCALE MISMATCH")
    print("=" * 60)

    stats = defaultdict(lambda: {'vals': []})

    for vid, feats in features_dict.items():
        for name in ['resnet', 'i3d_rgb', 'i3d_flow', 'concat_pca']:
            if name in feats:
                arr = feats[name].flatten()
                stats[name]['vals'].append(arr)

    print(f"\n{'Feature':<15} {'Mean':>10} {'Std':>10} {'Min':>10} {'Max':>10} {'L2 norm':>10}")
    print("-" * 65)
    for name, s in stats.items():
        all_vals = np.concatenate(s['vals'])
        # Per-frame L2 norm average
        all_frames = np.concatenate([feats[name] for feats in features_dict.values()
                                      if name in feats], axis=0)
        l2_norms = np.linalg.norm(all_frames, axis=1).mean()
        print(f"{name:<15} {all_vals.mean():>10.4f} {all_vals.std():>10.4f} "
              f"{all_vals.min():>10.4f} {all_vals.max():>10.4f} {l2_norms:>10.4f}")

    print("\n⚠️  WARNING THRESHOLDS:")
    print("  If Std(resnet) vs Std(i3d_rgb) differ by >5×  → normalize before concat")
    print("  If L2 norm(resnet) vs L2 norm(i3d_flow) differ by >3× → normalize")
    print("  If Max > 100 or Min < -100 → likely not normalized")


# ── 4. ZERO / DEAD FEATURES ─────────────────────────────────────────────────
def check_dead_features(features_dict):
    """
    Zero vectors = frames where I3D had no input (e.g. padding at start/end)
    or optical flow failed to compute.
    """
    print("\n" + "=" * 60)
    print("CHECK 4: ZERO VECTORS (dead frames)")
    print("=" * 60)

    for vid, feats in features_dict.items():
        for name in ['i3d_rgb', 'i3d_flow', 'concat_pca']:
            if name not in feats:
                continue
            arr = feats[name]
            zero_rows = np.where(np.all(arr == 0, axis=1))[0]
            if len(zero_rows) > 0:
                print(f"❌ {vid} / {name}: {len(zero_rows)} zero vectors at frames {zero_rows[:5]}...")
            
            # Also check near-zero (L2 norm < 1e-6)
            norms = np.linalg.norm(arr, axis=1)
            tiny = np.where(norms < 1e-6)[0]
            if len(tiny) > 0 and len(tiny) != len(zero_rows):
                print(f"⚠️  {vid} / {name}: {len(tiny)} near-zero vectors")

    print("  (Zero I3D flow vectors usually mean optical flow wasn't computed for that clip)")


# ── 5. TEMPORAL ALIGNMENT ───────────────────────────────────────────────────
def check_temporal_alignment(features_dict):
    """
    ResNet extracts 1 feature per frame.
    I3D extracts 1 feature per CLIP (usually 16 frames).
    If you simply matched them by index, they may be misaligned.
    """
    print("\n" + "=" * 60)
    print("CHECK 5: TEMPORAL ALIGNMENT")
    print("=" * 60)

    print("""
  ResNet:    [f1][f2][f3][f4][f5][f6][f7][f8] ...  (1 per frame)
  I3D:       [ clip 1  ][ clip 2  ][ clip 3  ] ...  (1 per 16 frames)

  If your I3D has N/16 vectors and ResNet has N vectors,
  you need to upsample I3D before concat:
  
      i3d_upsampled = np.repeat(i3d_features, 16, axis=0)[:N]
  
  OR downsample ResNet:
  
      resnet_downsampled = resnet[::16]  # take every 16th frame
    """)

    for vid, feats in features_dict.items():
        n_resnet = feats['resnet'].shape[0]
        n_rgb    = feats['i3d_rgb'].shape[0]
        n_flow   = feats['i3d_flow'].shape[0]
        ratio_rgb  = n_resnet / n_rgb  if n_rgb > 0 else 0
        ratio_flow = n_resnet / n_flow if n_flow > 0 else 0

        if abs(ratio_rgb - 1.0) > 0.1:
            print(f"⚠️  {vid}: ResNet has {n_resnet} frames, I3D RGB has {n_rgb} "
                  f"(ratio={ratio_rgb:.1f}×)")
        if abs(ratio_flow - 1.0) > 0.1:
            print(f"⚠️  {vid}: ResNet has {n_resnet} frames, I3D Flow has {n_flow} "
                  f"(ratio={ratio_flow:.1f}×)")


# ── 6. PCA DATA LEAKAGE CHECK ───────────────────────────────────────────────
def check_pca_leakage(pca_fit_indices, all_indices, split_test_indices):
    """
    PCA must be fit ONLY on training data.
    If you fit PCA on all data then split → data leakage.
    
    pca_fit_indices:  indices of videos used to fit PCA
    split_test_indices: indices of videos in test split
    """
    print("\n" + "=" * 60)
    print("CHECK 6: PCA DATA LEAKAGE")
    print("=" * 60)

    overlap = set(pca_fit_indices) & set(split_test_indices)
    if overlap:
        print(f"❌ LEAKAGE DETECTED: {len(overlap)} test videos were seen during PCA fitting!")
        print(f"   Affected videos: {overlap}")
        print("""
   FIX:
       from sklearn.decomposition import PCA
       
       # Fit PCA on TRAIN split only
       pca = PCA(n_components=512)
       pca.fit(train_features_concat)          # ← only train data
       
       # Transform both train and test
       train_pca = pca.transform(train_features_concat)
       test_pca  = pca.transform(test_features_concat)  # ← not fit here
        """)
    else:
        print("✓ No PCA leakage detected")


# ── 7. PCA VARIANCE EXPLAINED ───────────────────────────────────────────────
def check_pca_variance(pca_model):
    """
    Pass your fitted sklearn PCA model to check how much
    information is retained after compression.
    """
    print("\n" + "=" * 60)
    print("CHECK 7: PCA VARIANCE EXPLAINED")
    print("=" * 60)

    cumvar = np.cumsum(pca_model.explained_variance_ratio_)
    total_var = cumvar[-1]

    print(f"  Total variance retained with 512 components: {total_var*100:.1f}%")
    print(f"  Variance at 128 components: {cumvar[127]*100:.1f}%")
    print(f"  Variance at 256 components: {cumvar[255]*100:.1f}%")

    if total_var < 0.80:
        print(f"⚠️  Only {total_var*100:.1f}% variance retained — consider keeping more components")
    elif total_var > 0.95:
        print(f"✓ Good: {total_var*100:.1f}% variance retained")

    # Check if one feature type dominates
    # The first components of PCA will mostly capture the highest-variance source
    print("""
  TIP: To check which feature (ResNet vs I3D) dominates PCA:
  
      resnet_var  = pca.components_[:, :512]   # columns 0–511 = ResNet
      i3d_rgb_var = pca.components_[:, 512:1536]  # cols 512–1535 = I3D RGB  
      i3d_fl_var  = pca.components_[:, 1536:]  # cols 1536+ = I3D Flow
      
      print("ResNet contribution:", np.abs(resnet_var).mean())
      print("I3D RGB contribution:", np.abs(i3d_rgb_var).mean())
      print("I3D Flow contribution:", np.abs(i3d_fl_var).mean())
    """)


# ── 8. QUICK CORRELATION CHECK ──────────────────────────────────────────────
def check_feature_gt_correlation(features_dict):
    """
    Do high-importance frames actually have different feature norms
    than low-importance frames? If correlation is near 0, your
    features may not carry the signal needed for summarization.
    """
    print("\n" + "=" * 60)
    print("CHECK 8: FEATURE–GROUNDTRUTH CORRELATION")
    print("=" * 60)

    from scipy.stats import spearmanr

    correlations = defaultdict(list)

    for vid, feats in features_dict.items():
        gt = feats['gtscore']
        for name in ['resnet', 'i3d_rgb', 'i3d_flow', 'concat_pca']:
            if name not in feats:
                continue
            arr = feats[name]
            # Use L2 norm of feature as proxy for "activity"
            norms = np.linalg.norm(arr, axis=1)
            if len(norms) == len(gt):
                rho, _ = spearmanr(norms, gt)
                correlations[name].append(rho)

    print(f"\n{'Feature':<15} {'Mean Spearman ρ':>18} {'Std':>8}  Interpretation")
    print("-" * 65)
    for name, corrs in correlations.items():
        mean_r = np.mean(corrs)
        std_r  = np.std(corrs)
        if abs(mean_r) < 0.05:
            interp = "⚠️  near zero — feature norm carries little signal"
        elif abs(mean_r) < 0.15:
            interp = "→  weak signal"
        else:
            interp = "✓  useful signal"
        print(f"{name:<15} {mean_r:>18.4f} {std_r:>8.4f}  {interp}")


# ── MAIN ─────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    print("""
    ╔══════════════════════════════════════════════╗
    ║   STVT Feature Diagnostic Tool              ║
    ║   Load your features and call each check    ║
    ╚══════════════════════════════════════════════╝

    Usage:
        features_dict = load_your_features()   # see format in check_shapes()
        check_shapes(features_dict)
        check_nan_inf(features_dict)
        check_scales(features_dict)
        check_dead_features(features_dict)
        check_temporal_alignment(features_dict)
        check_pca_variance(your_fitted_pca_model)
        check_feature_gt_correlation(features_dict)
    """)
```

# Normalize + Concat + PCA for all 25 SumMe videos.
## Saves one .npy file per video: shape (N, 512) float32

```python
"""
Normalize + Concat + PCA for all 25 SumMe videos.
Saves one .npy file per video: shape (N, 512) float32
"""

import numpy as np
import os
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ── CONFIG ───────────────────────────────────────────────────────────────────
RGB_DIR      = rgb_path
FLOW_DIR     = flow_path
RESNET_DIR   = resnet_path
dataset='tvsum'
OUTPUT_DIR   = '/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/'+dataset+'/Features/normalized claude'
N_COMPONENTS = 512
# ─────────────────────────────────────────────────────────────────────────────

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ── STEP 1: Load all raw features, align per video ───────────────────────────
video_names = sorted([f.replace(".npy", "") for f in os.listdir(FLOW_DIR) if f.endswith(".npy")])
print(f"Found {len(video_names)} videos: {video_names[:3]}...")

raw = {}   # {vid: (resnet, rgb, flow)} all shape (N, D) — same N guaranteed

for vid in video_names:
    resnet = np.load(os.path.join(RESNET_DIR, f"{vid}.npy")).astype(np.float32)
    rgb    = np.load(os.path.join(RGB_DIR,    f"{vid}.npy")).astype(np.float32)
    flow   = np.load(os.path.join(FLOW_DIR,   f"{vid}.npy")).astype(np.float32)

    # Ensure 2D
    if resnet.ndim == 3: resnet = resnet.mean(axis=1)
    if rgb.ndim    == 3: rgb    = rgb.mean(axis=1)
    if flow.ndim   == 3: flow   = flow.mean(axis=1)

    # Use ResNet frame count as reference (it's 1 per frame)
    N = resnet.shape[0]

    # Upsample I3D to match ResNet frame count
    def upsample(arr, target):
        if arr.shape[0] == target:
            return arr
        # Use interpolation instead of repeat for smoother upsampling
        indices = np.linspace(0, arr.shape[0] - 1, target)
        lo = np.floor(indices).astype(int)
        hi = np.minimum(lo + 1, arr.shape[0] - 1)
        frac = (indices - lo)[:, None]
        return (arr[lo] * (1 - frac) + arr[hi] * frac).astype(np.float32)

    rgb  = upsample(rgb,  N)
    flow = upsample(flow, N)

    # Verify alignment
    assert resnet.shape[0] == rgb.shape[0] == flow.shape[0], \
        f"{vid}: shape mismatch resnet={resnet.shape[0]} rgb={rgb.shape[0]} flow={flow.shape[0]}"

    raw[vid] = (resnet, rgb, flow)
    print(f"  {vid}: N={N} | resnet={resnet.shape} rgb={rgb.shape} flow={flow.shape}")

# ── STEP 2: Fit scalers on ALL videos (per feature type) ─────────────────────
print("\nFitting StandardScaler...")

all_resnet = np.concatenate([raw[v][0] for v in video_names], axis=0)
all_rgb    = np.concatenate([raw[v][1] for v in video_names], axis=0)
all_flow   = np.concatenate([raw[v][2] for v in video_names], axis=0)

print(f"  Total frames: resnet={all_resnet.shape[0]}, rgb={all_rgb.shape[0]}, flow={all_flow.shape[0]}")
assert all_resnet.shape[0] == all_rgb.shape[0] == all_flow.shape[0], \
    "Frame count mismatch after alignment!"

sc_resnet = StandardScaler().fit(all_resnet)
sc_rgb    = StandardScaler().fit(all_rgb)
sc_flow   = StandardScaler().fit(all_flow)

# Check scales after normalization
resnet_n = sc_resnet.transform(all_resnet)
rgb_n    = sc_rgb.transform(all_rgb)
flow_n   = sc_flow.transform(all_flow)

print(f"\nAfter normalization:")
print(f"  {'Feature':<12} {'Std':>8}  {'L2 norm':>10}")
print(f"  {'-'*34}")
for name, arr in [("resnet", resnet_n), ("i3d_rgb", rgb_n), ("i3d_flow", flow_n)]:
    print(f"  {name:<12} {arr.std():>8.4f}  {np.linalg.norm(arr, axis=1).mean():>10.4f}")

# ── STEP 3: Fit PCA on normalized concatenated features ──────────────────────
all_concat = np.concatenate([resnet_n, rgb_n, flow_n], axis=1)  # (total_frames, 2560)
print(f"\nFitting PCA({N_COMPONENTS}) on [{all_concat.shape[0]} frames × {all_concat.shape[1]} dims]...")

pca = PCA(n_components=N_COMPONENTS, random_state=42)
pca.fit(all_concat)

var = pca.explained_variance_ratio_.sum()
print(f"Variance retained: {var*100:.1f}%")

c_resnet = np.abs(pca.components_[:, :512]).mean()
c_rgb    = np.abs(pca.components_[:, 512:1536]).mean()
c_flow   = np.abs(pca.components_[:, 1536:]).mean()
total    = c_resnet + c_rgb + c_flow
print(f"\nPCA contributions:")
print(f"  ResNet:   {c_resnet/total*100:.1f}%  (expected ~20%)")
print(f"  I3D RGB:  {c_rgb/total*100:.1f}%  (expected ~40%)")
print(f"  I3D Flow: {c_flow/total*100:.1f}%  (expected ~40%)")

# ── STEP 4: Transform and save per video ─────────────────────────────────────
print(f"\nSaving to {OUTPUT_DIR}/")

for vid in video_names:
    resnet, rgb, flow = raw[vid]

    resnet_norm = sc_resnet.transform(resnet)
    rgb_norm    = sc_rgb.transform(rgb)
    flow_norm   = sc_flow.transform(flow)

    concat  = np.concatenate([resnet_norm, rgb_norm, flow_norm], axis=1)
    pca_out = pca.transform(concat).astype(np.float32)

    out_path = os.path.join(OUTPUT_DIR, f"{vid}.npy")
    np.save(out_path, pca_out)
    print(f"  ✓ {vid}: {pca_out.shape}")

print(f"\nDone. {len(video_names)} files saved.")
print(f"Each file shape: (N_frames, {N_COMPONENTS}) float32")
```
