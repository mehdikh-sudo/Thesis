## Feature Bert Score

```python
import torch
import numpy as np

def feature_bertscore(candidate_feats, reference_feats):
    """
    Compute a BERTScore-like precision, recall, and F1 between candidate and reference frame features.
    Args:
        candidate_feats: (N_cand, D) torch.Tensor or np.array
        reference_feats: (N_ref, D) torch.Tensor or np.array
    """
    if isinstance(candidate_feats, np.ndarray):
        candidate_feats = torch.from_numpy(candidate_feats).float()
    if isinstance(reference_feats, np.ndarray):
        reference_feats = torch.from_numpy(reference_feats).float()

    # L2 normalize (like BERTScore does on embeddings)
    cand_norm = torch.nn.functional.normalize(candidate_feats, p=2, dim=1)
    ref_norm = torch.nn.functional.normalize(reference_feats, p=2, dim=1)

    # Cosine similarity matrix (N_ref x N_cand)
    sim = torch.matmul(ref_norm, cand_norm.T)  # (N_ref, N_cand)

    # Precision: for each candidate, find the best matching reference frame
    precision = sim.max(dim=0)[0].mean()  # max over rows (ref), average over candidates

    # Recall: for each reference, find best matching candidate frame
    recall = sim.max(dim=1)[0].mean()  # max over columns (cand), average over refs

    # F1
    f1 = 2 * precision * recall / (precision + recall + 1e-8)
    return precision.item(), recall.item(), f1.item()

# Example usage:
if __name__ == "__main__":
    # Example: 5 candidate frames, 7 reference frames, 1024-dim features
    feature=np.load('/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/SumMe/Features/I3d rgb-flow +Resnet (concat)-dim 512/Air_Force_One.npy')
    ref = feature[:5]
    cand=feature[:5]
    P, R, F1 = feature_bertscore(cand, ref)
    print(f"Feature BERTScore - P: {P:.4f}, R: {R:.4f}, F1: {F1:.4f}")
```

## Random splits for TVSum and SumMe datasets

```python
import random

def generate_random_splits(total_videos=50, num_splits=5, test_ratio=0.2):
    """
    Generate 5 random 80/20 splits for SumMe dataset
    """
    all_videos = list(range(1, total_videos + 1))  # [1, 2, ..., 25]
    splits = []
    
    for split_idx in range(num_splits):
        # Shuffle and split
        shuffled = all_videos.copy()
        random.shuffle(shuffled)
        
        test_size = int(len(shuffled) * test_ratio)  # 5 videos
        test_videos = sorted(shuffled[:test_size])
        train_videos = sorted(shuffled[test_size:])
        
        splits.append({
            'split': split_idx + 1,
            'train': train_videos,
            'test': test_videos,
            'test_str': ','.join(map(str, test_videos))
        })
        
        print(f"Split {split_idx + 1}:")
        print(f"  Train: {train_videos} ({len(train_videos)} videos)")
        print(f"  Test:  {test_videos} ({test_size} videos)")
        print(f"  Command: --test_dataset \"{splits[-1]['test_str']}\"")
        print()
    
    return splits

# Generate splits
splits = generate_random_splits(total_videos=50, num_splits=5, test_ratio=0.2)
```

```python
import h5py
import numpy as np
import pandas as pd
import os
dataset='TvSum'
stvt_Normalized_dataset_path='/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/'+dataset+'/H5 File/'+dataset+'_RFR_Normalized.h5'

out_h5_path='/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/'+dataset+'/H5 File/'+dataset+'_RFR_Normalized_gtscore.h5'

gtscore_dir='/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/'+dataset+'/gtscores_05/'


with h5py.File(stvt_Normalized_dataset_path, "r") as src, h5py.File(out_h5_path, "w") as dst:
    for vid in src.keys():
        print("video: ",src[vid])
        grp_src = src[vid]
        grp_dst = dst.create_group(vid)
        
        print(src[vid].keys())
        

        # Replace features with rfr Normalized npy
        video_name = mapping_tvsum.get(vid, None)
        print("video_name: ",video_name)
        
        # print(grp_src['change_points'][:])

        gtscore_path = os.path.join(gtscore_dir, video_name)
       
        print(gtscore_path)
        
        if not os.path.exists(gtscore_path):
           print(f"⚠️ feature file not found for {vid}: {gtscore_path}")
           continue

        gtscore = np.load(gtscore_path)
    


        grp_dst.create_dataset('change_points',data=grp_src['change_points'])
        
        grp_dst.create_dataset("feature", data=grp_src['feature'])
        grp_dst.create_dataset("label",data=grp_src['label'])
        grp_dst.create_dataset("length",data=grp_src['length'])

        grp_dst.create_dataset('n_frame_per_seg',data=grp_src['n_frame_per_seg'])
        grp_dst.create_dataset('user_summary',data=grp_src['user_summary'])

        grp_dst.create_dataset('gtscore',data=gtscore)
        



        
```
