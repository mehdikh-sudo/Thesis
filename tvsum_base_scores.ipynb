```python
import os
out_h5_path='/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/TVsum/h5 File/TvSum_Rgb_Flow_resnet_matched_05.h5'
scores_files_root='/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/TvSum/Scores_05/'
gtscores_files_root='/Users/mehdikhosravi/Master/Thesis/STVT-main/Feature extraction/TvSum/gtscores_05/'
mapping=mapping_tvsum
with h5py.File('/Users/mehdikhosravi/Master/Thesis/STVT-main/STVT/datasets/datasets/TVSum_Rgb_Flow_Resnet.h5', "r") as src, h5py.File(out_h5_path, "w") as dst:
    for vid in src.keys():
        print(vid)
        grp_src = src[vid]
        grp_dst = dst.create_group(vid)
        print(dst.keys())

        # Replace scores with new 0-5 scores

        file_name = mapping.get(vid, None)

        print(F'video_number: ',file_name)

        if file_name is None:
            print(f"⚠️ No mapping found for {vid}, skipping features")
            continue

        score_path = os.path.join(scores_files_root, file_name)
        gtscore_path=os.path.join(gtscores_files_root, file_name)

        print(score_path)

        if not os.path.exists(score_path):
           print(f"⚠️ feature file not found for {vid}: {score_path}")
           continue
        
        score = np.load(score_path)
        score=np.transpose(score)
        gtscore=np.load(gtscore_path)

        # feat=l2_normalize(feat)
        
        grp_dst.create_dataset('change_points',data=grp_src['change_points'])
        
        grp_dst.create_dataset("feature", data=grp_src['feature'])

        grp_dst.create_dataset("label",data=gtscore)

        grp_dst.create_dataset("length",data=grp_src['length'])

        grp_dst.create_dataset('n_frame_per_seg',data=grp_src['n_frame_per_seg'])

        grp_dst.create_dataset('user_summary',data=score)
        
```
