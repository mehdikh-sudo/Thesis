```python
import os
import pandas as pd
```

## Function:

```python
def avg_test_result(result_path):
    dfs = []
    for file in os.listdir(result_path):
        if file.endswith(".csv"):
            file_path = os.path.join(result_path, file)
            df = pd.read_csv(file_path)
            dfs.append(df)
    print(f"Loaded {len(dfs)} CSV files")


    avg_F_measure=0
    avg_val_loss=0
    avg_train_loss=0
    avg_spearman=0
    avg_kendal=0
    avg_best_F_measure=0
    avg_bert_score=0
    min_fscore=100
    test_datasets=[]
    
    for df in dfs:
        test_datasets.append(df['test_dataset'].iloc[-1])
        avg_F_measure+=df['F_measure_k'].iloc[-1]
        avg_best_F_measure+=df['Best_F_measure'].iloc[-1]
        
        avg_val_loss+=df['val_loss'].iloc[-1]
        avg_train_loss+=df['train_loss'].iloc[-1]
        avg_spearman+=df['Spearman_rho_k'].iloc[-1]
        avg_kendal+=df['Kendall_tau_k'].iloc[-1]
        avg_bert_score+=df['BERTScore_k'].iloc[-1]

        if df['F_measure_k'].iloc[-1]< min_fscore:
            min_fscore=df['F_measure_k'].iloc[-1]
            

    
    avg_train_loss/=5
    avg_val_loss/=5
    avg_F_measure/=5
    avg_best_F_measure/=5
    avg_spearman/=5
    avg_kendal/=5
    avg_bert_score/=5

    results = {
    "test_dataset": "Average 5 fold ="+str(test_datasets),
    "Batch_size": 40,
    "train_loss":avg_train_loss,
    "val_loss": avg_val_loss,

    "F_measure_k": avg_F_measure,
    "Best_F_measure": avg_best_F_measure,

    "Spearman_rho_k": avg_spearman,
    "Kendall_tau_k": avg_kendal,
    "BERTScore_k": avg_bert_score,
}

    df = pd.DataFrame([results])

    # Save to CSV
    df.to_csv(result_path+"/cross_validation_avg_results.csv", index=False)

    return ('avg_F:',avg_F_measure,'avg_vall_loss:',avg_val_loss,'avg_train_loss:',avg_train_loss,'Min Fscore',min_fscore)

```
